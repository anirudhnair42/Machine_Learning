```python
#gathering the my user data through the spotipy library using their username
client_credentials_manager = SpotifyClientCredentials(client_id='6fc48df8020e4a68ba511ecc57304b1f', client_secret='8b9ebf5b2f08431790e6a7cb38d402dc')

def show_tracks(tracks):
    for i, item in enumerate(tracks['items']):
        track = item['track']
        print (" %d %s %s" % (i, track['artists'][0]['name'],track['name']))
        
def get_track_features(track_id,sp):
    if track_id is None:
        return None
    else:
        features = sp.audio_features([track_id])
    return features

def get_features(tracks,sp):
    tracks_with_features=[]
    for track in tracks:
        features = get_track_features(track['id'],sp)
        print (track['name'])
        if not features:
            print("passing track %s" % track['name'])
            pass
        else:
            f = features[0]
            tracks_with_features.append(dict(name=track['name'],artist=track['artist'],id=track['id'],danceability=f['danceability'],energy=f['energy'],loudness=f['loudness'],speechiness=f['speechiness'],acousticness=f['acousticness'],tempo=f['tempo'],liveness=f['liveness'],valence=f['valence']))
        #time.sleep(0.1)
    #print(tracks_with_features[0])
    return tracks_with_features

def get_tracks_from_playlists(username, sp):
    playlists = sp.user_playlists(username)
    trackList = []
    for playlist in playlists['items']:
        if playlist['owner']['id'] == username:
            print (playlist['name'],' no. of tracks: ',playlist['tracks']['total'])
            results = sp.user_playlist(username, playlist['id'],fields="tracks,next")
            tracks = results['tracks']
            for i, item in enumerate(tracks['items']):
                track = item['track']
                trackList.append(dict(name=track['name'], id=track['id'], artist=track['artists'][0]['name']))
    # print(trackList[0])
    return trackList

def write_to_csv(track_features):
    df = pd.DataFrame(track_features)
    df.drop_duplicates(subset=['name','artist'])
    print ('Total tracks in data set', len(df))
    df.to_csv('mySongsDataset.csv',index=False)
    
def main(username):
    sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
    print ("Getting user tracks from playlists")
    tracks = get_tracks_from_playlists(username, sp)
    print ("Getting track audio features")
    tracks_with_features = get_features(tracks,sp)
    print ("Storing into csv")
    write_to_csv(tracks_with_features)
    
if __name__ == '__main__':
    main('anirudhnair4242')
```

Spotify stores the following data for all the songs in its database:
1. A confidence measure from 0.0 to 1.0 of whether the track is acoustic. 1.0 represents high confidence
the track is acoustic. Looking at the songs in my playlist, it looks like a majority of my liked songs
measure low on acousticness. This is because most of the songs I listen to are electronic in nature and
thus, they end up not having a lot of acousticness.
2. Danceability describes how suitable a track is for dancing based on a combination of musical elements
including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable
and 1.0 is most danceable. My Music taste mostly include songs that are very danceable, as the
histogram for danceability is skewed to the right.
3. Energy: Energy is a measure from 0.0 to 1.0 and represents a perceptual measure of intensity and
activity. Typically, energetic tracks feel fast, loud, and noisy. For example, death metal has high energy,
while a Bach prelude scores low on the scale. Perceptual features contributing to this attribute include
dynamic range, perceived loudness, timbre, onset rate, and general entropy. My taste is very neuutral in
terms of energy as the distribution seems to be uniform.
4. Loudness: The overall loudness of a track in decibels (dB). Loudness values are averaged across the
entire track and are useful for comparing relative loudness of tracks. Loudness is the quality of a sound
that is the primary psychological correlate of physical strength (amplitude). Values typical range between
-60 and 0 db, however I have scaled it between 0 and 1 for the purposes of classification. This shows
that most of my songs are loud in nature as the histogram is skewed towards the right.
5. Speechiness detects the presence of spoken words in a track. The more exclusively speech-like the
recording (e.g. talk show, audio book, poetry), the closer to 1.0 the attribute value. Values above 0.66
describe tracks that are probably made entirely of spoken words. Values between 0.33 and 0.66
describe tracks that may contain both music and speech, either in sections or layered, including such
cases as rap music. Values below 0.33 most likely represent music and other non-speech-like tracks.
My histogram in this case is skewed to the left, but that there is a significant number of songs between
0.33 and 0.66. But since my taste is mostly electronic, it is understandable that I prefer songs below the
0.33 spectrum.
6. Tempo: The overall estimated tempo of a track in beats per minute (BPM). In musical terminology,
tempo is the speed or pace of a given piece and derives directly from the average beat duration. Since
the values are normalized, the value of the highest frequency appears to be in the middle of the
spectrum, which I would imagine to be around 120 beats per minute. However, the histogram shows
that I prefer slower songs more.
7. Valence: A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track. Tracks
with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence
sound more negative (e.g. sad, depressed, angry). It is interesting to note that my preference is towards
the lower values, but I assure you I am not a sad or depressed person

# Using unsupervised learning to cluster songs

The K-Means clustering algorithm is a great way to find underlying distributions in data. Since I am dealing
with numerous songs with no label on the type of environment I would want to listen to them, I can use the K
means clusterning algorithm to discover underlying patterns in my music and use it to automate the process
of song classificsation. I first plot the number of clusters against the Sum of Squared distances of each point
from the cluster mean to understand how much variation do different levels of clusters account for. Through
the plot, we can see that going for more than 3 clusters does no account for a lot of variation, and thus the
dataset can be best classified into three classes

```python
ssq = []
k = range(1,15)
for i in k:
 kmeans = KMeans (n_clusters = i)
 kmeans = kmeans.fit(features)
 ssq.append(kmeans.inertia_)
```

```python
plt.plot(k, ssq, '-o', color='black')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of squared distances from the cluster mean')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(features)
```

Since we are dealing with multiple features in the dataset, we cannot plot all the features in a single graph as
that would be impossible. We thus use Principle Component Analysis (PCA) to reduce dimensionality by
focusing on variation between different features and creating axes that account for the most amount of
variation (PC1>PC2). We do not use LDA because the LDA model assumes that each of the features when
plotted normally aganst each other would remain to be normally to be normally distributed. However as seen
from the histograms above, we are dealing with features that are skewed in different directions. LDA also
assumes the variance of different features remain constant across classes, which is not the case here.

```python
y = kmeans.predict(features)
pca = PCA(n_components=2)
pc = pca.fit_transform(features)
```

```python
components = pd.DataFrame(pc)
components['label'] = y
components.columns = ['x','y','label']
clusters = sns.lmplot(data=components, x='x', y='y', hue='label', fit_reg=False, legend=True, legend_out=True)
```

# Training the classification models

```python
X = features
y = y
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)

#building the classifier for the data
logit = LogisticRegression(solver = "saga")
logit.fit(X_train, y_train)
```

```python
ytest_pred = logit.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logit.score(X_test, y_test)))
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
# Train the model using the training sets
knn.fit(X_train,y_train)
```

```python
ytrain_pred = knn.predict(X_test)
print('Accuracy of K Neighbors Classifier classifier on train set: {:.2f}'.format(knn.score(X_test, y_test)))
```

```python
#Import svm model
from sklearn import svm
#Create a svm Classifier
svml = svm.SVC(kernel="linear")
#Train the model using the training sets
svml.fit(X_train, y_train)
#Predict the response for test dataset
svml_pred = svml.predict(X_test)
print('Accuracy of Support Vector Machine classifier on train set: {:.2f}'.format(svml.score(X_test, y_test)))
```

```python
#Import svm model
from sklearn import svm
#Create a svm Classifier
svm = svm.SVC(kernel="poly")
#Train the model using the training sets
svm.fit(X_train, y_train)
#Predict the response for test dataset
svm_pred = svm.predict(X_test)
print('Accuracy of Support Vector Machine classifier on train set: {:.2f}'.format(svm.score(X_test, y_test)))
```

```python
#Import svm model
from sklearn import svm
#Create a svm Classifier
svm = svm.SVC(kernel="rbf")
#Train the model using the training sets
svm.fit(X_train, y_train)
#Predict the response for test dataset
svm_pred = svm.predict(X_test)
print('Accuracy of Support Vector Machine classifier on train set: {:.2f}'.format(svm.score(X_test, y_test)))
```

```python
mlp = MLPClassifier()
mlp.fit(X_train, y_train)
mlp_pred = svm.predict(X_test)
print('Accuracy of Multilayer Perceptron (Neural Net) classifier on train set: {:.2f}'.format(mlp.score(X_test, y_test)))
```
